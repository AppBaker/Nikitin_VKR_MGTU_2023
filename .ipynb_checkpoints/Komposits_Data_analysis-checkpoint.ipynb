## Анализ датасета



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('datasets/raw_join_data.csv', index_col="Unnamed: 0")
```

```python
df.head()
```

```python
df.info()
```

Проверим какие свойства можно преобразовать в категориальные

```python
property_cols = {}
for col in df.columns:
    if len(df[col].unique()) < 10:
        property_cols[col] = f"{len(df[col].unique())} - значения"

print(property_cols)
        
```

Значения угла нашивки 

```python
angle_degree = pd.get_dummies(df["Угол нашивки, град"], prefix="angle_degree")
```

```python
df["Угол нашивки, град"].value_counts()
```

Меняем данные угла нашивки. Примем 0 за 0 градусов, 1 за 90 градусов

```python
df["Угол нашивки, град"] = df["Угол нашивки, град"].replace({90:1})
```
